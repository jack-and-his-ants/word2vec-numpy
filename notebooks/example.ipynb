```python
import sys
import os
sys.path.append(os.path.abspath(".."))
import numpy as np
from datasets import load_dataset
from word2vec import Word2VecModel
from word2vec import Vocabulary
```

```python
model = Word2VecModel.load('../word2vec/stable_model/model1.npz')
vocab = Vocabulary.load_dict('../word2vec/stable_model/vocab1.pkl')
```

Finding similar words

```python
# Examples for which the model works well
example_words = ["mother","america","north","king","officer",'book','creation']
for word in example_words:
    model_response = model.get_k_closest_neighbours(vocab.get_index_by_word(word),4)
    print(f"Similar words to {word} are: {(vocab.get_word_by_index(model_response))}")
    print("")
```

```python
# Examples where the model does not work well
example_words = ["cat","hair","suit","ground","table",'bag','wisdom']
for word in example_words:
    model_response = model.get_k_closest_neighbours(vocab.get_index_by_word(word),4)
    print(f"Similar words to {word} are: {(vocab.get_word_by_index(model_response))}")
    print("")
```

```python
# detecting analogies
# correct examples
analogy_set = [
    ("king","man","woman"),
    ("boy","man","woman"),
    ('girl','woman','man'),
    ('actor','man','woman'),
    ('running','run','walk'),
    ("paris", "france","germany"),
    ('father','man','woman'),
]
for word1,word2,word3 in analogy_set:
    response = model.analogy(word1,word2,word3,vocab,1)
    print(f"{word1} - {word2} + {word3} => {response[0]}")
```

```python
# incorrect examples
analogy_set = [
    ("queen","woman","man",'king'),
    ("berlin","germany","france","paris"),
    ('son','man','woman',"daughter"),
    ('husband','man','woman','wife'),
    ('walking','walk','run','running'),
    ("princess", "woman","man",'prince')

]
for word1,word2,word3,expected in analogy_set:
    response = model.analogy(word1,word2,word3,vocab,1)
    print(f"{word1} - {word2} + {word3} => {response[0]}, expected: {expected}")
```
